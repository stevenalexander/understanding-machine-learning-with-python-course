# Understanding machine learning course

## Workflow

### Asking the right question

Need a statement to direct and validate work
- define end goal, starting point, and how to achieve goal
- define scope
- define target performance
- define context for usage
- define how solution will be created

### Preparing your data

- Find the data we need
- Inspect and clean the data
- Explore the data
- Mold the data to Tidy data

Tidy datasets are easy to manipulate, model and visualize, and have a specific structure.
- each variable is a column
- each observation is a row
- each type of observational unit is a table

#### Getting data

- Google (often untrustworthy)
- Government databases
- Professional or company data sources
- Your company
- Your department
- All of the above

Data rules
- Closer the data is to what you are predicting, the better.
- Data will never be in the format you need
- Accurately predicting rare events is difficult
- Track how you manipulate the data (for reproduction)

Correlated columns - same information in a different format
- add little information
- can cause algorithms to get confused

If the case you are looking for is rare in the data you may not be able to use some techniques (e.g. less than 30% of data).

Missing data
- common problem
- options
  - ignore
  - drop observations (rows)
  - replace values (imputing)

Imputing options
- replace with mean/median
- replace with expert knowledge derived value

### Selecting your algorithm

- Over 50 in sci-kit

Factors affecting selection:
- learning type
  - prediction (supervised/unsupervised)
- result

- complexity

- basic vs enhanced
  - enhanced are variations of basic

    - additional functionality

    - less complex
    - easier to understand

### Training the model

Letting specific data teach a machine learning algorithm to create a specific prediction model.

Retrain?
- new data for better predictions
- verifiy training performance with new data

Split the data into training (70%) and test data (30%).

Train model with training data, evaluate model with test data.

Selecting training features
- minimum features

### Testing the models accuracy

Use a confusion matrix to get
- TP (true positive)
- FP (false positive)
- FN (false negative)
- TN (true negative)

Classification report tells you:
- recall ( TP / (TP + FN) ) - higher the better
- precision ( TP / (TP + FP) ) - higher the better

### Performance Improvements options

- Adjust current algorithm
- Get more data or improve data
- Improve training
- Switch algorithms

#### New algorithm - Random Forest

- Ensemble algorithm
- Fits multiple decision trees with subsets of data
- Averages tree results to improve performance and control overfitting

#### Fixing overfitting

Overfitting is when the algorithm learns the training data too well.

- Regularization hyperparameters
- Cross validation
- Bias - variance trade-off
  - sacrifice some perfection for better overall performance

#### Unbalanced classes

When there are more of one class than the other, e.g. our diabetes data, 65% no diabetes, 35% diabetes.
Can cause biases estimation yielding poor prediction results.

#### Cross validation

K-fold Cross Validation
- split data into folds (sets of data)
- repeat training with all folds but one acting as validation data
- for each fold
  - determine best hyperparameter value
  - then set model hyperparameter value to average best
  
Algorithm + Cross validation = AlgorithmCV
- runs the algorithm K times
- can be used like normal algorithm
